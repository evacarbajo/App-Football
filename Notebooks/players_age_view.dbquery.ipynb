```sql
CREATE OR REPLACE VIEW players_age_view AS
SELECT
    player_id,
    CASE 
        WHEN month(current_date()) > month(date_of_birth)
          OR (
              month(current_date()) = month(date_of_birth)
              AND dayofmonth(current_date()) >= dayofmonth(date_of_birth)
          )
        THEN year(current_date()) - year(date_of_birth)
        ELSE year(current_date()) - year(date_of_birth) - 1
    END AS age
FROM football.`silver-football-data`.players_silver;
```
