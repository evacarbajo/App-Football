```python
from pyspark.sql import SparkSession

#Leer CSV y guardar una tabla bronze 
def create_bronze_table(csv_path: str, table_name: str):
    
    # Obtener la sesi√≥n de Spark
    spark = SparkSession.builder.getOrCreate()

    # Leer CSV e inferir esquema
    bronze_table = (spark.read
        .format("csv") 
        .option("header", "true") 
        .option("inferSchema", "true") 
        .load(csv_path)
    )
    
    # Guardar como tabla Bronze en Delta
    bronze_table.write \
        .format("delta") \
        .mode("overwrite") \
        .saveAsTable(f"football.`bronze-football-data`.{table_name}")

    #.option("overwriteSchema", "true")
    

create_bronze_table("/Volumes/football/football-data/football-csv/appearances.csv", "appearances_bronze")
create_bronze_table("/Volumes/football/football-data/football-csv/club_games.csv", "club_games_bronze")
create_bronze_table("/Volumes/football/football-data/football-csv/clubs.csv", "clubs_bronze")
create_bronze_table("/Volumes/football/football-data/football-csv/competitions.csv", "competitions_bronze")
create_bronze_table("/Volumes/football/football-data/football-csv/game_events.csv", "game_events_bronze")
create_bronze_table("/Volumes/football/football-data/football-csv/game_lineups.csv", "game_lineups_bronze")
create_bronze_table("/Volumes/football/football-data/football-csv/games.csv", "games_bronze")
create_bronze_table("/Volumes/football/football-data/football-csv/player_valuations.csv", "player_valuations_bronze")
create_bronze_table("/Volumes/football/football-data/football-csv/players.csv", "players_bronze")
create_bronze_table("/Volumes/football/football-data/football-csv/transfers.csv", "transfers_bronze")


```
